## Python-Based Solution (Linear Programming)

### What this means:

scipy is a scientific Python library

optimize is a module inside scipy

linprog is a linear programming solver

👉 This line tells Python:

“I want to use the linear programming solver.”

### Writing the Constraints in Python Form
Mathematical constraints:
X1 ≥ 500
X2 ≥ 600 

linprog only accepts ≤ (less-than-or-equal-to) constraints.

So we convert:
X1 ≥ 500 =>  -X1 ≤ -500
X2 ≥ 600 =>  -X2 ≤ -600 

This represents:
Constraint

| Constraint | x1 | x2 |
| ---------- | -- | -- |
| −x1 ≤ −500 | −1 | 0  |
| −x2 ≤ −600 | 0  | −1 |

Each row = one constraint

Each column = one variable


b = [-500, -600]
This corresponds to:

-X1 ≤ -500

-X2 ≤ -600


```python
# Importing the Optimization Tool
from scipy.optimize import linprog    

# Objective function coefficients
c = [10, 10]  # x1, x2                          # How Python sees it: c[0] = 10 → coefficient of x1
## This represents the objective function                           # c[1] = 10 → coefficient of x2
# Minimize Z = 10 X1 + 10 X2
# x1 = production in Month 1
# x2 = production in Month 2
# Cost per unit in each month = $10 


# Constraints (x1 ≥ 500, x2 ≥ 600)
A = [[-1, 0],
     [0, -1]]
b = [-500, -600]

# Non-negativity bounds
bounds = [(0, None), (0, None)]           # What this means: x1 ≥ 0 /  x2 ≥ 0 / None means no upper limit  X1 ≥ 0, X2 ≥ 0

result = linprog(c, A_ub=A, b_ub=b, bounds=bounds)    # This line does everything: “Find values of x1 and x2 that: 
                                                      # Minimize 10x1 + 10x2
                                                      # Satisfy all constraints
                                                      # Respect non-negativity”
                                                      # c Objective function
                                                      # A_ub  Left side of ≤ constraints
                                                      # b_ub  Right side of ≤ constraints
                                                      # bounds Variable limits


print(result.x)        # Optimal Decision Values
print(result.fun)      # Minimum Cost Value
```


## Python Output

Month 1 production = 500

Month 2 production = 600

Minimum cost = 11,000  Meaning  Z=10(500)+10(600)=11,000
