## Python_based Solution (Linear Programming)

```python
# Import the linear programming solver from the SciPy optimization library
from scipy.optimize import linprog

# Define the objective function coefficients
# This represents: Minimize Z = 10*x1 + 10*x2
# where:
# x1 = production quantity in Month 1
# x2 = production quantity in Month 2
c = [10, 10]

# Define the constraint coefficient matrix (A_ub)
# Each row represents one constraint
# Since linprog only supports <= constraints,
# we convert:
# x1 >= 500  -->  -x1 <= -500
# x2 >= 600  -->  -x2 <= -600
A = [
    [-1, 0],   # Coefficients for constraint: -x1 <= -500
    [0, -1]    # Coefficients for constraint: -x2 <= -600
]

# Define the right-hand side values for the constraints (b_ub)
# These correspond to the converted demand constraints
b = [
    -500,      # Right-hand side for constraint: -x1 <= -500
    -600       # Right-hand side for constraint: -x2 <= -600
]

# Define bounds for each decision variable
# (0, None) means the variable must be >= 0 and has no upper limit
# This enforces non-negativity:
# x1 >= 0
# x2 >= 0
bounds = [
    (0, None), # Bounds for x1
    (0, None)  # Bounds for x2
]

# Call the linear programming solver
# This line finds the values of x1 and x2 that:
# - Minimize the objective function (c)
# - Satisfy all inequality constraints (A_ub * x <= b_ub)
# - Respect the variable bounds
result = linprog(
    c,          # Objective function coefficients
    A_ub=A,     # Constraint coefficient matrix
    b_ub=b,     # Constraint right-hand side values
    bounds=bounds
)

# Print the optimal values of the decision variables
# result.x contains the values of x1 and x2
print(result.x)

# Print the minimum value of the objective function
# result.fun contains the optimal total cost
print(result.fun)
```
